### You can also create a dataset from a GIF using the following code.

```python
# importing Image class from PIL package
from PIL import Image, ImageSequence

import cv2
import numpy as np
import imageio
#from imageio import

# # Open the GIF file
gif_path = '/content/gdrive/MyDrive/julia-2.gif'
gif = Image.open(gif_path)

# Extract individual frames from the GIF.
frames = [frame.copy() for frame in ImageSequence.Iterator(gif)]

for i, frame in enumerate(frames):
    png_path = f'/content/julia/img_{i}.png'
    frame.save(png_path, format='PNG')
```
